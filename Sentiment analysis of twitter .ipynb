```python
#append both the dataset i.e. train and test dataset so that we can do cleaning in one go
combi=data.append(data1,ignore_index=True)
combi
```

```python
#remove the pattern using a function   
def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for i in r:
        input_txt = re.sub(i, '', input_txt)
        
    return input_txt   
```

```python
#remove @ using regular expression
combi['tidy_tweet'] = np.vectorize(remove_pattern)(combi['tweet'], "@[\w]*")
```

```python
combi
```

```python
#removing all the special character except # as it adds value in knowing whether tweet is racist or not
combi['tidy_tweet']=combi['tidy_tweet'].str.replace("[^a-zA-Z#]"," ")
combi
```

```python
#removing the 3 letter word as it does not add any values to the dataset (letter like the,in,out)
combi['tidy_tweet'] = combi['tidy_tweet'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
```
